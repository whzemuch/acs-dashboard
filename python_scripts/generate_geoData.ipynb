```python
import geopandas as gpd
import pandas as pd
import numpy as np

# # 1. Load counties shapefile
# url = "https://www2.census.gov/geo/tiger/TIGER2022/COUNTY/tl_2022_us_county.zip"
# counties_gdf = gpd.read_file(url)[["GEOID", "NAME", "STATEFP", "geometry"]]

# 2. Select 10 counties per state
selected = (
    counties_gdf.groupby("STATEFP")
    .apply(lambda x: x.sample(min(10, len(x)), random_state=42), include_groups=False)
    .reset_index(drop=True)
)

# 3. Generate migration flows (random demo data)
years = [2018, 2019, 2020]
records = []
np.random.seed(42)

for _, row in selected.iterrows():
    for y in years:
        n_flows = np.random.randint(2, 6)
        for _ in range(n_flows):
            dest = selected.sample(1).iloc[0]
            while dest["GEOID"] == row["GEOID"]:
                dest = selected.sample(1).iloc[0]

            if np.random.rand() > 0.5:
                origin_geoid, dest_geoid = row["GEOID"], dest["GEOID"]
            else:
                origin_geoid, dest_geoid = dest["GEOID"], row["GEOID"]

            flow_count = np.random.randint(10, 1001)
            records.append({
                "year": y,
                "origin_geoid": origin_geoid,
                "dest_geoid": dest_geoid,
                "flow": flow_count
            })

migration_df = pd.DataFrame(records)

# 4. Attach real county names
migration_df = migration_df.merge(
    counties_gdf[["GEOID", "NAME"]],
    left_on="origin_geoid", right_on="GEOID",
    how="left"
).rename(columns={"NAME": "origin_name"}).drop(columns=["GEOID"])

migration_df = migration_df.merge(
    counties_gdf[["GEOID", "NAME"]],
    left_on="dest_geoid", right_on="GEOID",
    how="left"
).rename(columns={"NAME": "dest_name"}).drop(columns=["GEOID"])





```

```python
migration_df.head()
```

```python
# Calculate net migration = inflow - outflow per county per year

# Inflow: sum of flows where county is destination
inflow = (
    migration_df.groupby(["year", "dest_geoid"])["flow"]
    .sum()
    .reset_index()
    .rename(columns={"dest_geoid": "GEOID", "flow": "inflow"})
)

# Outflow: sum of flows where county is origin
outflow = (
    migration_df.groupby(["year", "origin_geoid"])["flow"]
    .sum()
    .reset_index()
    .rename(columns={"origin_geoid": "GEOID", "flow": "outflow"})
)

# Merge inflow and outflow
net_mig = pd.merge(inflow, outflow, on=["year", "GEOID"], how="outer").fillna(0)

# Calculate net migration
net_mig["net_migration"] = net_mig["inflow"] - net_mig["outflow"]




```

```python
# 1. Reproject to an equal-area CRS (US Albers, EPSG:5070)
counties_proj = counties_gdf.to_crs("EPSG:5070")

# 2. Compute centroids in projected CRS
centroids = counties_proj.geometry.centroid

# 3. Convert back to WGS84 (EPSG:4326) for lon/lat
centroids = gpd.GeoSeries(centroids, crs="EPSG:5070").to_crs("EPSG:4326")

# 4. Attach to counties_gdf
counties_gdf["lon"] = centroids.x
counties_gdf["lat"] = centroids.y
```

```python
counties_gdf.head()
```

```python
centroids_df = counties_gdf[["GEOID", "NAME", "lon", "lat"]]
```

```python
migration_df = migration_df \
    .merge(centroids_df, left_on="origin_geoid", right_on="GEOID", how="left", suffixes=("", "_origin")) \
    .merge(centroids_df, left_on="dest_geoid", right_on="GEOID", how="left", suffixes=("", "_dest"))

# Rename columns
migration_df = migration_df.rename(columns={
    "NAME": "origin_name",
    "lon": "origin_lon",
    "lat": "origin_lat",
    "NAME_dest": "dest_name",
    "lon_dest": "dest_lon",
    "lat_dest": "dest_lat"
}).drop(columns=["GEOID", "GEOID_dest"])
```
